```python
from scipy.stats import bootstrap

means = [np.mean(scores_MCTS_Highest),np.mean(scores_MCTS_Lowest),np.mean(scores_MCTS_Random),np.mean(scores_MCTS_Simple),np.mean(scores_MCTS_Smart)]
result = bootstrap((scores_MCTS_Highest,),statistic=np.mean,n_resamples=99999)
ci_55 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_55}")

result = bootstrap((scores_MCTS_Lowest,),statistic=np.mean,n_resamples=99999)
ci_12 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_12}")

result = bootstrap((scores_MCTS_Random,),statistic=np.mean,n_resamples=99999)
ci_8 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_8}")

result = bootstrap((scores_MCTS_Simple,),statistic=np.mean,n_resamples=99999)
ci_4 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_4}")

result = bootstrap((scores_MCTS_Smart,),statistic=np.mean,n_resamples=99999)
ci_5 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_5}")

x_values = np.arange(len(means))
# Confidence intervals for the y-axis
cis = [ci_55, ci_12, ci_8, ci_4,ci_5]

# Plot the confidence intervals
plt.figure(figsize=(8, 6))

for i, ci in enumerate(cis):
    plt.plot([i, i], ci, marker='_', color='grey', markersize=8, linewidth=2)

plt.scatter(x_values, means, color='black', marker='o')

plt.xticks(x_values, ['Highest Agent', 'Lowest Agent', 'Random Agent', 'Simple Agent', 'Smart Agent'])
plt.xlabel('Agent Type')
plt.ylabel('Average Score')
plt.axhline(y=6.5, color='blue', linestyle='--', label='Avg = 6.5')
plt.title('Performance of MCTS against Agent Types')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_simple = pd.read_csv('results/MCTSAgent-SimpleAgent.csv')
df_random = pd.read_csv('results/MCTSAgent-RandomAgent.csv')
df_highest = pd.read_csv('results/MCTSAgent-HighestAgent.csv')
df_lowest = pd.read_csv('results/MCTSAgent-LowestAgent.csv')
df_smart = pd.read_csv('results/MCTSAgent-SmartAgent.csv')

df_simple['Agent'] = 'Simple'
df_random['Agent'] = 'Random'
df_highest['Agent'] = 'Highest'
df_lowest['Agent'] = 'Lowest'
df_smart['Agent'] = 'Smart'

df_combined_agents = pd.concat([df_simple, df_random, df_highest, df_lowest, df_smart])

mean_scores_agents = df_combined_agents.groupby('Agent')['MCTSAgent'].mean().reset_index()
for i, row in mean_scores_agents.iterrows():
    print(f"Mean for playing against {row['Agent']}: {row['MCTSAgent']:.2f}")

plt.figure(figsize=(10, 6))
sns.barplot(x='Agent', y='MCTSAgent', data=mean_scores_agents, ci=None, palette='viridis')
plt.title('Average MCTSAgent Scores for Different Agents')
plt.xlabel('Agent')
plt.ylabel('Average MCTSAgent Score')
plt.show()
```
