```python

import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

df_depth_55 = pd.read_csv('results/MCTSAgent_Smart-SmartAgent.csv')
df_depth_4 = pd.read_csv('results/MCTSAgent_Smart-SmartAgent-EXP_2-MAX_DEPTH_4.csv')
df_depth_8 = pd.read_csv('results/MCTSAgent_Smart-SmartAgent-EXP_2-MAX_DEPTH_8.csv')
df_depth_12 = pd.read_csv('results/MCTSAgent_Smart-SmartAgent-EXP_2-MAX_DEPTH_12.csv')

df_combined_depth = pd.concat([
    df_depth_55.assign(MAX_DEPTH=55),
    df_depth_4.assign(MAX_DEPTH=4),
    df_depth_8.assign(MAX_DEPTH=8),
    df_depth_12.assign(MAX_DEPTH=12)
])

mean_scores = df_combined_depth.groupby('MAX_DEPTH')['MCTSAgent_Smart'].mean().reset_index()

for depth_value in df_combined_depth['MAX_DEPTH'].unique():
    subset_depth = df_combined_depth[df_combined_depth['MAX_DEPTH'] == depth_value]
    mean_score_depth = subset_depth['MCTSAgent_Smart'].mean()
    variance_score_depth = subset_depth['MCTSAgent_Smart'].var()
    print(f"Mean for MAX_DEPTH = {depth_value}: {mean_score_depth:.2f}")
    print(f"Variance for MAX_DEPTH = {depth_value}: {variance_score_depth:.2f}")
    print("-------------------")


t_stat, p_value = ttest_ind(df_depth_55['MCTSAgent_Smart'], df_depth_4['MCTSAgent_Smart'])
print(f"T-test between MAX_DEPTH=55 and MAX_DEPTH=4: p-value = {p_value:.4f}")

if p_value < 0.05:
    print(f"Reject the null hypothesis that MCTS with MAX_DEPTH=55 performs the same on average as MCTS MCTS with MAX_DEPTH=4(p-value < 0.05)")
else:
    print(f"Accept the null hypothesis that MCTS with MAX_DEPTH=55 performs the same on average as MCTS with MAX_DEPTH=4 (p-value >= 0.05)")

plt.plot(mean_scores['MAX_DEPTH'], mean_scores['MCTSAgent_Smart'], marker='o')
plt.title('Mean Scores vs. MAX_DEPTH')
plt.xlabel('MAX_DEPTH')
plt.ylabel('Mean Scores')
plt.show()
```

```python
# Bootstrapping to see whether that changes something for the accuracy... not really

from scipy.stats import bootstrap

means = [np.mean(df_depth_55['MCTSAgent_Smart']),np.mean(df_depth_12['MCTSAgent_Smart']),np.mean(df_depth_8['MCTSAgent_Smart']),np.mean(df_depth_4['MCTSAgent_Smart'])]
result = bootstrap((df_depth_55['MCTSAgent_Smart'],),statistic=np.mean,n_resamples=99999)
ci_55 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_55}")

result = bootstrap((df_depth_12['MCTSAgent_Smart'],),statistic=np.mean,n_resamples=99999)
ci_12 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_12}")

result = bootstrap((df_depth_8['MCTSAgent_Smart'],),statistic=np.mean,n_resamples=99999)
ci_8 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_8}")

result = bootstrap((df_depth_4['MCTSAgent_Smart'],),statistic=np.mean,n_resamples=99999)
ci_4 = result.confidence_interval
print(f"Bootstrap Confidence Interval of the Mean: {ci_4}")

x_values = np.arange(4)
# Confidence intervals for the y-axis
cis = [ci_55, ci_12, ci_8, ci_4]

# Plot the confidence intervals
plt.figure(figsize=(8, 6))

for i, ci in enumerate(cis):
    plt.plot([i, i], ci, marker='_', markersize=8, linewidth=2, label=f'CI {i+1}')

plt.scatter(x_values, means, color='red', marker='o', label='Means')

plt.xticks(x_values, ['No limit', 'Depth 12', 'Depth 8', 'Depth 4'])
plt.xlabel('Confidence Intervals')
plt.ylabel('Values')
plt.title('Impact of Change in Maximum Depth of MCTS')
plt.legend()
plt.grid(True)
plt.show()
```
